```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img1 = cv2.imread(r'D:\pythonTest\image\norm.jpg')
img2 = cv2.imread(r'D:\pythonTest\image\orb.jpg')

alpha = 0.1
blended1 = (img1 * alpha) + (img2 * (1-alpha))              # 1
blended1 = blended1.astype(np.uint8) #  
blended2 = cv2.addWeighted(img1, alpha, img2, (1-alpha), 0) # 2

cv2.imwrite(r'D:\pythonTest\image\result\re1.jpg',blended1)

for i, img in enumerate([img1,img2,blended1, blended2]):
    plt.subplot(2,2,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img[:, :, ::-1])
plt.show()
```
